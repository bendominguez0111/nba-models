```python
import sys
import os

base_dir = os.path.dirname(os.path.abspath(''))
sys.path.insert(0, base_dir)

from model.models.threes import run_threes_model
```

```python
from nba_api.stats.static import teams
```

```python
team_abrvs = [team['abbreviation'] for team in teams.get_teams()]
team_abrvs = [abrv for abrv in team_abrvs if abrv != "BOS"]
team_abrvs[:5]
```

```python
player_name = 'Jayson Tatum'
sim_results = {}
for opponent in team_abrvs:
    simmed_fgms = run_threes_model(player_name=player_name, opponent=opponent, n_simulated_games=10_000, plot=False)
    sim_results[opponent] = simmed_fgms
```

```python
import pandas as pd
```

```python
pd.DataFrame(
        data=sim_results,
        index = [f'simulation_{i}' for i in range(1, 10_001)]
).to_csv('sim_results.csv')
```
